```python
def sumhashtotarget(nums, target):
#hash map of seen numbers
  peeked_number = {}

#iterate over the index and values over a hash map of an array
  for index, value in enumerate(nums):
      complement = target - value
      if complement in peeked_number:
        #return index of the complement seen in peeked_num
        return [peeked_number[complement], index]
      # if not found on this pass add the current value to the peeked number hash
      peeked_number[value] = index
```

The Hash Map Trigger: Whenever a problem requires you to know the exact counts, frequencies, or quantities of items, and explicitly tells you that order does not matter (like an anagram), your brain should immediately yell HASH MAP! Hash maps are the ultimate "inventory" data structure. Anytime a problem feels like taking inventory or tallying votes, a Hash Map is almost always the answer.

```python
s = "anagram"
t = "nagaram"
```

```python
s = "rat"
t = "car"
```

```python
def isAnagramHASH(s,t):
  #anagram = contains same characters
  if len(s) != len(t):
    return False

  ctS = {}
  ctT = {}

  for let in s:
    if let in ctS:
      ctS[let] += 1
      print(ctS)
    else:
      ctS[let] =1
      print(ctS)

  for let in t:
    if let in ctT:
      ctT[let] +=1
      print(ctT)
    else:
      ctT[let] =1
      print(ctT)

  return ctS == ctT

```

```python
isAnagramHASH(s,t)
```
