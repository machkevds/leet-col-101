```python
def sumhashtotarget(nums, target):
#hash map of seen numbers
  peeked_number = {}

#iterate over the index and values over a hash map of an array
  for index, value in enumerate(nums):
      complement = target - value
      if complement in peeked_number:
        #return index of the complement seen in peeked_num
        return [peeked_number[complement], index]
      # if not found on this pass add the current value to the peeked number hash
      peeked_number[value] = index
```

```python
def twosumhapra(nums, target):

  seen = {}

  for index, value in enumerate(nums):
    com = target - value
    if com in seen:
      return [seen[com], index]
    seen[value] = index
```

```python
def pointerApproach(nums, target):
    left = 0
    right = len(nums) - 1

    while left < right:
        current_sum = nums[left] + nums[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:  # current_sum > target
            right -= 1
```
